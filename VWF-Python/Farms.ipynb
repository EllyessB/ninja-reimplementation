```python
def define_farm_region(wf, margin=10):
    lonBounds = np.asarray([[np.min(wf.lon)-margin],[np.max(wf.lon)+margin]])
    lonBounds = np.clip(lonBounds, -180, 180)
    latBounds = np.asarray([[np.min(wf.lat)-margin],[np.max(wf.lat)+margin]])
    latBounds = np.clip(latBounds, -90, 90)
    bounds = {"lat": latBounds, "lon": lonBounds}
    return bounds

define_farm_region(wf, margin=10)
```
